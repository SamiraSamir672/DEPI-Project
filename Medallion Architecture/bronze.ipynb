# Bronze Layer - Data Understanding (Raw Layer)

## Objective
The Bronze layer is responsible for ingesting the raw Bitcoin datasets from the source without applying any cleaning or transformations. This layer preserves the original data and serves as the foundation for the remaining stages of the Medallion Architecture.

## Steps Performed
- Download the Bitcoin historical dataset from Kaggle.
- Load the four different timeframes:
  - 15 Minutes
  - 1 Hour
  - 4 Hours
  - 1 Day
- Display the first few rows of each dataset.
- Inspect the number of rows and columns.
- Review the available column names.
- Check data types.
## Output
- **Input:** Raw CSV files from Kaggle
- **Output:** "bronze_15m.csv"
              "bronze_1h.csv"
              "bronze_4h.csv"
              "bronze_1d.csv"

> **Note:** No data cleaning, preprocessing, or feature engineering is performed in the Bronze layer. The data is stored exactly as received from the source.

```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#download data from kaggle
import kagglehub
path = kagglehub.dataset_download("novandraanugrah/bitcoin-historical-datasets-2018-2024")
```

```python
# Load raw datasets
df_15m = pd.read_csv(path + "/btc_15m_data_2018_to_2025.csv")
df_1h  = pd.read_csv(path + "/btc_1h_data_2018_to_2025.csv")
df_4h  = pd.read_csv(path + "/btc_4h_data_2018_to_2025.csv")
df_1d  = pd.read_csv(path + "/btc_1d_data_2018_to_2025.csv")
```

```python
# Save Bronze datasets

df_15m.to_csv("bronze_15m.csv", index=False)
df_1h.to_csv("bronze_1h.csv", index=False)
df_4h.to_csv("bronze_4h.csv", index=False)
df_1d.to_csv("bronze_1d.csv", index=False)  
```
